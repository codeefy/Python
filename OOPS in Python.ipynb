**Creating a class ATM**

```python
#pascal case 
# class StudentDetails: two words start with capital letter
class Atm:
    #constructor : it is a special function which is called at the time of object creation
    def __init__(self): 
        self.pin=''
        self.balance=0
       # print("execute")
        self.menu()
# There are two types of classes in python inbuilt like list , tuple, dict 
# and user defied classes this is user defined classes . 
    def menu(self):
        user_input=input("""HI how can i help you 
                         1. Press 1 to create pin
                         2. Press 2 to change Pin 
                         3. Press 3 to check Balance
                         4. Press 4 to withdraw
                         5 anything else to exit  """)
        if user_input=='1':
            #create Pin 
            self.Create_pin()
        elif user_input=='2':
            #change pin
            self.change_pin()
        elif user_input=='3':
            #check balance 
            self.check_balance()
        elif user_input=='4':
            #withdraw
            self.withdraw()
            
        else:
            exit()

    def Create_pin(self):
        user_pin=input("Enter your pin ")
        self.pin=user_pin

        user_balance=int(input("enter Your balance "))
        self.balance=user_balance
        print("your pin is created successfully")
        self.menu()
    
    def change_pin(self):
        old_pin=input("enter old pin")

        if old_pin==self.pin:
            #let him change the pin 
            new_pin=input("enter New pin")
            self.pin=new_pin
            print("pin change succesfully")
            self.menu()
        else:
            print("Permissioon Denied")
            self.menu()
    
    def check_balance(self):
        user_pin=input("enter your Pin")
        if user_pin==self.pin:
            print("your balance is ",self.balance)
        else:
            print("permission denied")
            self.menu()
    
    def withdraw(self):
        user_pin=input("enter your Pin")
        if user_pin==self.pin:
            amount=int(input("enter amount to withdraw "))
            if amount<=self.balance:
                self.balance=self.balance - amount
                print("please collect your cash")
                print("your remaining balance is ",self.balance)
                self.menu()
            else:
                print("insufficient balance")
                self.menu()
        else:
            print("permission denied")
            self.menu()




    

```

**Fraction**


```python
class fraction:

    #parameterized constructor 
    def __init__(self,x,y):
        self.num=x
        self.den=y
    def __str__(self):
        return '{}/{}'.format(self.num,self.den)
    
    def __add__(self,other):
        new_num=self.num*other.den+other.num*self.den # basic fraction addition cross multiplication 
        new_den=self.den*other.den
        return '{}/{}'.format(new_num,new_den)
    def __sub__(self,other):
        new_num=self.num*other.den-other.num*self.den 
        new_den=self.den*other.den
        return '{}/{}'.format(new_num,new_den)
    def __mul__(self,other):
        new_num=self.num*other.num
        new_den=self.den*other.den
        return '{}/{}'.format(new_num,new_den)
    def __truediv__(self,other):
        new_num=self.num*other.den
        new_den=self.den*other.num
        return '{}/{}'.format(new_num,new_den)

```

```python
fr1=fraction(3,4)
fr2=fraction(1,2)
```

```python
#print(fr1,fr2)
#print(fr1+fr2) # it will give error because we have not defined how to add two fraction objects 
print(fr1+fr2) # now it will add up  becaaues we have defined the how we add two fraction object 
print(fr1-fr2)
print(fr1*fr2)
print(fr1/fr2)
```
